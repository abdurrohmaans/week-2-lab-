## Section 5: Simple Perceptron Prediction

**What this section is about:**  
This section combines dot product, bias, and activation into a single prediction function.  
This is a complete single-neuron model.

image.png

```python
import numpy as np

x = np.array([0.5, 0.2, 0.1])  # misol uchun 3 ta input

w = np.array([0.4, 0.3, 0.9])

b = 0.1

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def neuron_predict(x, w, b):
    z = np.dot(x, w) + b
    y = sigmoid(z)
    return y

output = neuron_predict(x, w, b)
print("go:", output)
```

### ✏️ Exercise 5
Change the values of `w` and `b`.  
Observe how the prediction changes.

```python
import numpy as np

def step_function(x):
    return 1 if x >= 0 else 0

def predict(x, w, b):
    z = np.dot(x, w) + b
    return step_function(z)

x = np.array([1, 4])

# First set of parameters
w = np.array([2, -1])
b = 0
print(predict(x, w, b))

# Change weights and bias
w = np.array([1, 1])
b = -3
print(predict(x, w, b))

# Another change
w = np.array([-1, 2])
b = 0
print(predict(x, w, b))

```

## Section 7: Decision Boundary Preview

**What this section is about:**  
A perceptron learns a straight-line decision boundary.  
This section previews how weights and bias define that line.

```python
import numpy as np
import matplotlib.pyplot as plt

w1, w2 = 2, -1
b = -1

def plot_decision_boundary(w1, w2, b):

    x1 = np.linspace(-5, 5, 100)

    x2 = -(w1*x1 + b) / w2

    plt.plot(x1, x2, 'r-', label='Decision boundary')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.title('Perceptron Decision Boundary')
    plt.grid(True)
    plt.legend()
    plt.xlim(-5,5)
    plt.ylim(-5,5)
    plt.show()



```
